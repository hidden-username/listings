#### Listed by ZipCode for range Jan 2010 - July 2018

##### *Note: Not all ZipCodes supported.  

```python
import pandas as pd
from matplotlib import pyplot as plt
from ipywidgets import widgets
from IPython.display import display

csv = "/data/InventoryMeasure_Zip_Public.csv"
df = pd.read_csv(csv, index_col="RegionName")
df = df.T
```

```python
def plotZip(zipcode):
    y = df[zipcode].values[6:108]
    x = df.index[6:108]
    plt.title(df[zipcode][3] + " ZipCode: " + str(zipcode))
    plt.plot(x, y)
```

```python
# Example Plot for Dallas
plotZip(75070)
```

```python
text = widgets.Text()
display(text)

def handle_submit(sender):
    plotZip(int(text.value))

text.on_submit(handle_submit)
```
